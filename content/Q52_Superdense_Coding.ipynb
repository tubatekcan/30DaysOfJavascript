## Süperyoğun Kodlama Algoritması

Ayça Bora'ya $b_1b_0$ şeklindeki 2-bitlik bir mesaj göndermek istiyor. Burada $b_i \in \{0,1\}$ değerlerinden ancak birini alabilen klasik bilgi bitleridir. Ayça'nın göndermek isteyebileceği muhtemel bilgi çiftleri $00$, $01$, $10$ veya $11$ değerlerinden birisidir.

Ayça $b_1b_0$ mesajını elindeki kubite aşağıdaki yöntem ile kodlayabilir:
 - Eğer $b_0$'ın değeri 1 ise, Ayça kendi kubitine Z kapısı uygulasın.
 - Eğer $b_1$'ın değeri 1 ise, Ayça kendi kubitine X kapısı uygulasın.

Ayça bu yönergeyi izleyerek göndermek istediği klasik mesajı kendi kubitine kodladıktan sonra bu kubiti Bora'ya "postalıyor".
Bora, Ayça'nın yolladığı kubiti aldıktan sonra sırayla şu işlemleri yapsın:
 - Ayça'den gelen kubitin kontrol, kendi kubitinin hedef olduğu CNOT kapısı uygula.
 - Ayça'dan gelen kubite Hadamard uygula.
 - Ayça'nın gönderdiği kubiti ölç, bu ölçümün sonucu $b_0$ bilgisini verecek.
 - Kendi kubitini ölç, bu ölçümün sonucu $b_1$ bilgisini verecek.

İddia edilene göre Bora, bu şekilde Ayça'nın gönderdiği $b_1b_0$ mesajını almış olur.

## Protokolün Ayrıntılı İncelemesi

image.png

İncelemeye dolaşık kubit durumundan itibaren başlayacağız. Dolaşık kubitlerle ilgili ayrıntılı tartışmalar önceki defterlerde yapılmıştı. Durumlar yazılırken Qiskit'in kullandığı kubit sırası dikkate alındı.

 - Dolaşık kubit durumu:
 $$
 \ket{\Psi_0} = \frac{1}{\sqrt{2}}(\ket{00}+\ket{11})
 $$
Bu kubiterden sağdaki, yani indisi 0 olan kubit, Ayça'nın ve soldaki, yani indisi 1 olan kubit ise Bora'nın kubitidir.

Şimdi Ayça göndermek istediği klasik mesajı çeşitli kuantum kapıları uygulayarak kendi kubite şifreleyecek. Burada Ayça yapması gereken işlemlere göndermek istediği klasik mesaja göre karar veriyor.
 - Şifreleme aşaması:
  - $b_0$ klasik bitine göre yapılan işlemin sonucu:
    - Eğer $b_0=0$ ise, $$\ket{\Psi_1} = \ket{\Psi_0} = \frac{1}{\sqrt{2}}(\ket{00}+\ket{11})$$
    - Eğer $b_0=1$ ise, $$\ket{\Psi_1} = (I\otimes Z)\ket{\Psi_0} = \frac{1}{\sqrt{2}}(\ket{00}-\ket{11})$$

  - $b_1$ klasik bitine göre yapılan işlemin sonucu:
    - Eğer $b_0=0$ ve $b_1=0$ ise, $$\ket{\Psi_2} = \ket{\Psi_1} = \frac{1}{\sqrt{2}}(\ket{00}+\ket{11})$$
    - Eğer $b_0=0$ ve $b_1=1$ ise, $$\ket{\Psi_2} = (I\otimes X)\ket{\Psi_1} = (I\otimes X)\frac{1}{\sqrt{2}}(\ket{00}+\ket{11}) = \frac{1}{\sqrt{2}}(\ket{01}+\ket{10})$$
    - Eğer $b_0=1$ ve $b_1=0$ ise, $$\ket{\Psi_2} = \ket{\Psi_1} = \frac{1}{\sqrt{2}}(\ket{00}-\ket{11})$$
    - Eğer $b_0=1$ ve $b_1=1$ ise, $$\ket{\Psi_2} = (I\otimes X) \ket{\Psi_1} = (I\otimes X)\frac{1}{\sqrt{2}}(\ket{00}-\ket{11}) = \frac{1}{\sqrt{2}}(\ket{01}-\ket{10})$$


$\ket{\Psi_2}$ durumunu özetleyelim:

$$
\ket{\Psi_2}=\left\{
    \begin{array}{ll}
    \frac{1}{\sqrt{2}}(\ket{00}+\ket{11}), & \text{eğer } b_1b_0 = 00 \\
    \frac{1}{\sqrt{2}}(\ket{00}-\ket{11}), & \text{eğer } b_1b_0 = 01 \\
    \frac{1}{\sqrt{2}}(\ket{01}+\ket{10}), & \text{eğer } b_1b_0 = 10 \\
    \frac{1}{\sqrt{2}}(\ket{01}-\ket{10}), & \text{eğer } b_1b_0 = 11
    \end{array}
    \right.
$$

Sonra Ayça elindeki kubiti Bora'ya gönderiyor. Bu gönderme işlemi ancak ve ancak uygun bir kuantum kanalı ile gerçekleşeceğinden dolayı, bu bilgi taşıması sırasında fizik yasalarını çiğneyen bir durum olması söz konusu değildir. Bora, Ayça'nın gönderdiği kubiti teslim aldığında deşifreleme işlemini uyguluyor. Bu işlemi inceleyelim.

 - Deşifrelemenin ilk adımı: CNOT kapısı:

$$
\begin{align}
\ket{\Psi_3}&=\text{CNOT}_{01}\ket{\Psi_2}\\
            &=\text{CNOT}_{01}\left\{
                \begin{array}{ll}
                \frac{1}{\sqrt{2}}(\ket{00}+\ket{11}), & \text{eğer } b_1b_0 = 00 \\
                \frac{1}{\sqrt{2}}(\ket{00}-\ket{11}), & \text{eğer } b_1b_0 = 01 \\
                \frac{1}{\sqrt{2}}(\ket{01}+\ket{10}), & \text{eğer } b_1b_0 = 10 \\
                \frac{1}{\sqrt{2}}(\ket{01}-\ket{10}), & \text{eğer } b_1b_0 = 11
                \end{array}
                \right. \\\\
            &=\left\{
                \begin{array}{ll}
                \frac{1}{\sqrt{2}}\text{CNOT}_{01}(\ket{00}+\ket{11}), & \text{eğer } b_1b_0 = 00 \\
                \frac{1}{\sqrt{2}}\text{CNOT}_{01}(\ket{00}-\ket{11}), & \text{eğer } b_1b_0 = 01 \\
                \frac{1}{\sqrt{2}}\text{CNOT}_{01}(\ket{01}+\ket{10}), & \text{eğer } b_1b_0 = 10 \\
                \frac{1}{\sqrt{2}}\text{CNOT}_{01}(\ket{01}-\ket{10}), & \text{eğer } b_1b_0 = 11
                \end{array}
                \right. \\\\
            &=\left\{
                \begin{array}{ll}
                \frac{1}{\sqrt{2}}(\ket{00}+\ket{01}), & \text{eğer } b_1b_0 = 00 \\
                \frac{1}{\sqrt{2}}(\ket{00}-\ket{01}), & \text{eğer } b_1b_0 = 01 \\
                \frac{1}{\sqrt{2}}(\ket{11}+\ket{10}), & \text{eğer } b_1b_0 = 10 \\
                \frac{1}{\sqrt{2}}(\ket{11}-\ket{10}), & \text{eğer } b_1b_0 = 11
                \end{array}
                \right. \\\\
            &=\left\{
                \begin{array}{ll}
                \frac{1}{\sqrt{2}}\ket{0}(\ket{0}+\ket{1}), & \text{eğer } b_1b_0 = 00 \\
                \frac{1}{\sqrt{2}}\ket{0}(\ket{0}-\ket{1}), & \text{eğer } b_1b_0 = 01 \\
                \frac{1}{\sqrt{2}}\ket{1}(\ket{1}+\ket{0}), & \text{eğer } b_1b_0 = 10 \\
                \frac{1}{\sqrt{2}}\ket{1}(\ket{1}-\ket{0}), & \text{eğer } b_1b_0 = 11
                \end{array}
                \right. \\\\
            &=\left\{
                \begin{array}{ll}
                \ket{0}\ket{+}, & \text{eğer } b_1b_0 = 00 \\
                \ket{0}\ket{-}, & \text{eğer } b_1b_0 = 01 \\
                \ket{1}\ket{+}, & \text{eğer } b_1b_0 = 10 \\
                \ket{1}\ket{-}, & \text{eğer } b_1b_0 = 11
                \end{array}
                \right. \\
\end{align}
$$

 - Deşifrelemenin ikinci adımı: Hadamard kapısı:

$$
\begin{align}
\ket{\Psi_4} &= (I\otimes H)\ket{\Psi_3}\\\\
             &=(I\otimes H)\left\{
                \begin{array}{ll}
                \ket{0}\ket{+}, & \text{eğer } b_1b_0 = 00 \\
                \ket{0}\ket{-}, & \text{eğer } b_1b_0 = 01 \\
                \ket{1}\ket{+}, & \text{eğer } b_1b_0 = 10 \\
                \ket{1}\ket{-}, & \text{eğer } b_1b_0 = 11
                \end{array}
                \right. \\\\
             &=\left\{
                \begin{array}{ll}
                \ket{0}H\ket{+}, & \text{eğer } b_1b_0 = 00 \\
                \ket{0}H\ket{-}, & \text{eğer } b_1b_0 = 01 \\
                \ket{1}H\ket{+}, & \text{eğer } b_1b_0 = 10 \\
                \ket{1}H\ket{-}, & \text{eğer } b_1b_0 = 11
                \end{array}
                \right. \\
\end{align}
$$

_Hatırlatma_: $H\ket{+}=\ket{0}$ ve $H\ket{-}=\ket{1}$ idi.

$$
\begin{align}
\ket{\Psi_4} &=\left\{
                \begin{array}{ll}
                \ket{0}\ket{0}, & \text{eğer } b_1b_0 = 00 \\
                \ket{0}\ket{1}, & \text{eğer } b_1b_0 = 01 \\
                \ket{1}\ket{0}, & \text{eğer } b_1b_0 = 10 \\
                \ket{1}\ket{1}, & \text{eğer } b_1b_0 = 11
                \end{array}
                \right. \\
\end{align}
$$

Bu ifadeden açıkça görülmektedir ki $\ket{\Psi_4}$ üzerine uygulanacak olan ölçüm sonucunda ilgili $b_1b_0$ bilgileri %100 olasılıkla belirecektir.
